**Harris Corner Detector (HCD)**

Lý thuyết của Harris Corner Detector (HCD)

Góc là các vùng trong ảnh có sự thay đổi lớn về cường độ theo mọi hướng. HCD dựa vào sự biến đổi cường độ sáng mà phát hiện đặc trưng góc, ta có một cửa sổ trượt trên ảnh khi cửa sổ phát hiện ra sự thay đổi cường độ sáng ở mọi hướng và sự thay đổi đó lớn hơn một ngưỡng nhất định thì đó là góc.

Khi dịch chuyển một cửa sổ nhỏ qua các vùng của ảnh, ta thấy:

- Mặt phẳng : cửa sổ không có sự thay đổi về cường độ khi di chuyển theo bất kì hướng nào.
- Cạnh : cửa sổ chỉ thay đổi khi di chuyển theo chiều ngang
- Góc : cửa sổ thay đổi khi di chuyển bất kì hướng nào

Trong đó:

- w(x,y): cửa sổ trượt tại tọa độ (x,y)
- I(x+u,y+v): cường độ tại tọa độ đã dịch chuyển một khoảng (u,v)
- I(x,y): cường độ tại tọa độ điểm hiện tại
- E(u,v): sự thay đổi cường độ với cửa sổ (x,y) so với cường độ với cửa sổ (x+u, y+v)
